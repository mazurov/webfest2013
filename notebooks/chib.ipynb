```python
import rootnotes
from lib import utils
```

```python
cfg = utils.json("configs/chib1s.json")
```

```python
import ROOT
tuples = ROOT.TChain("ChibAlg/Chib")
tuples.Add(cfg["tuples2011"])
```

```python
cut = cfg["cut"]
cut["pt_ups"] = (18,30)
cut["dm"] = (0.31, 1.43)
```

```python
from lib.chib import ChibModel
from lib import fit
```

```python
canvas=rootnotes.default_canvas()
model = ChibModel(canvas,
                  dm_begin=cut["dm"][0],
                  dm_end=cut["dm"][1],
                  nbins=57,
                  bgorder=2,
                  frac=[0.6, 0.5, 0.5]
                  )
```

```python
f = fit.Fit(model=model,
            tuples=tuples,
            cut=cut,
            field="dm",
            is_unbinned=False,
            nbins=57,
            has_splot=False)
```

```python
result = f.process()
```
